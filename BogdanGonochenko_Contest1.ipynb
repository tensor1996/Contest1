```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv("train.csv")

train_df.head()
```

```python
bad_list = []
for sub_df in train_df.groupby('url'):
    
    if ( sub_df[1].target.count() > 2 ):
        if ( all(sub_df[1].target) ):
            
            bad_list.append(sub_df[0])
```

```python
len(bad_list)
```

```python
test_df = pd.read_csv("test.csv")

test_df.head()
```

```python
from sklearn.metrics import f1_score
```

```python
X_train = train_df["url"].values+' '+train_df["title"].values
X_test = test_df["url"].values+' '+test_df["title"].values
y_train = train_df["target"].astype(int).values
```

# ML baseline

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression

vectorizer = CountVectorizer()
model = LogisticRegression(C=755.0)
```

```python
X_train_vectorized = vectorizer.fit_transform(X_train)
```

```python
%%time
model.fit(X_train_vectorized, y_train)

```

```python
y_pred = model.predict(X_train_vectorized)
```

```python
f1_score(y_train, y_pred)
```

### Submit

```python
X_test_vectorized = vectorizer.transform(X_test)

test_df["target"] = model.predict(X_test_vectorized).astype(bool)
```

```python
test_df['in_bad_list'] = test_df['url'].map(lambda x: 1 if x in bad_list else 0)
```

```python
test_df[test_df.target]
```

```python
test_df['target'] = test_df.apply(lambda x: x['target'] or bool(x['in_bad_list']) , axis = 1)
```

```python
test_df['target'] = test_df.apply(lambda x: ( x['target']) or ('porn' in x['url']) or ('XXX' in x['url']) or 
                               ('xxx' in x['url']) or ('cuckold' in x['title']), axis = 1)
```
